## 4.FIRST VISUALIZATION

### 4.1.TOP 10 STATES BY ADOPTION PERCENTAGE

this graph shows the top 10 states by using adoption percentage

```python
top_adoption = df.sort_values(
    by='Adoption_percentage',
    ascending=False
).head(10)
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    top_adoption['States'],
    top_adoption['Adoption_percentage']
)

plt.xticks(rotation=45)

plt.title("Top 10 States by Adoption Percentage")

plt.xlabel("States")

plt.ylabel("Adoption Percentage")

plt.show()
```

Observation:
States with higher adoption percentages demonstrate stronger Panchayat digital governance implementation and planning participation.

### 4.2.ONLINE PAYMENT ANALYSIS

this graph shows digital finance governace adoptation chart.

```python
top_payment = df.sort_values(
    by='Gram_Panchayat_With_Online_Payment',
    ascending=False
).head(10)
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    top_payment['States'],
    top_payment['Gram_Panchayat_With_Online_Payment']
)

plt.xticks(rotation=45)

plt.title("Top States with Online Payment Adoption")

plt.xlabel("States")

plt.ylabel("Online Payment Count")

plt.show()
```

### 4.3.GOVERNANCE ACTIVITY ANALYSIS

ths graph shows about the goveranance activity by collected Total numer of Panchayat data as it measures overall governance participation.

```python
top_activity = df.sort_values(
    by='Total_number(sum)',
    ascending=False
).head(10)
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    top_activity['States'],
    top_activity['Total_number(sum)']
)

plt.xticks(rotation=45)

plt.title("Top States by Governance Activity")

plt.xlabel("States")

plt.ylabel("Governance Activity Score")

plt.show()
```

## 5.HEATMAP

heatmap is usually used to show the relationship strength between variables

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True
)

plt.title("Correlation Heatmap of Governance Indicators")

plt.show()
```

## 6.PIE CHART

this graph shows the actual proportional contribuion.

```python
top5 = top_adoption.head(5)

plt.figure(figsize=(8,8))

plt.pie(
    top5['Adoption_percentage'],
    labels=top5['States'],
    autopct='%1.1f%%'
)

plt.title("Top 5 States by Adoption Percentage")

plt.show()
```

Key Insights:
- Several states demonstrated strong Panchayat digitization participation.
- Online payment adoption varied significantly across states.
- Governance activity indicators showed positive relationships with planning uploads.
- Data visualization improved comparative understanding of governance performance.
