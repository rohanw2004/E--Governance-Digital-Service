## 1.Importing Required Libraries

In this step, the required Python libraries are imported for performing basic data analysis and visualization. Pandas is used for data handling and manipulation, while Matplotlib is used for graphical representation of service usage patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2.Creating Sample Dataset

Since real-time government usage data is not publicly available, a sample dataset has been created to simulate service usage within the UMANG App ecosystem. The dataset represents approximate user interaction across different government services.

```python
data = {
    "Service": [
        "EPFO",
        "DigiLocker",
        "Aadhaar",
        "Bill Payments",
        "Farmer Schemes",
        "Certificates",
        "Job Services"
    ],

    "Users": [
        4200,
        3900,
        5100,
        2800,
        2200,
        3100,
        2600
    ]
}
```

## 3.Converting Data into Structured Format

The collected sample data is converted into a DataFrame structure to organize the information into rows and columns. This allows easier processing, analysis, and visualization.

```python
df = pd.DataFrame(data)

df
```

## 4.Performing Basic Statistical Analysis

This step calculates simple statistics from the dataset such as total users and average usage. These values provide an initial understanding of service adoption and usage patterns.

```python
print("Dataset Information")
print("\nTotal Users:")
print(df["Users"].sum())
print("\nAverage Usage:")
print(df["Users"].mean())
```

## 5.Identifying Most Frequently Used Service

The objective of this step is to determine which government service has the highest user interaction within the sample dataset. Such analysis can help identify services that receive greater public engagement.

```python
most_used = df.loc[df["Users"].idxmax()]

print("Most Used Service:")
print(most_used)
```

## 6.Visualizing Service Usage Data

Data visualization helps transform numerical information into graphical insights. A bar chart is created to compare service usage patterns and understand which categories receive higher user participation.

```python
plt.figure(figsize=(8,5))
df.plot(x="Service", y="Users", kind="bar", color="skyblue")
plt.title("Usage of Government Services")
plt.xlabel("Services")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 8.Conclusion

This basic analysis demonstrates how service usage data can be analyzed for digital governance platforms.

Such analysis can help identify popular services, understand user behavior, and support better decision-making in e-governance systems.
