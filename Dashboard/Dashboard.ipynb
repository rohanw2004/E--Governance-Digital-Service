# Dashboard KPI Section

```python
print("Total States:",df["States"].nunique())
print("Average Adoption Percentage:",
round(df["Adoption_percentage"].mean(),2))
print("Total Governance Activity:",
df["Total_number(sum)"].sum())
```

# Adoption Dashboard Chart

```python
plt.figure(figsize=(12,6))

top=df.sort_values(
    "Adoption_percentage",
    ascending=False
).head(10)

plt.bar(
    top["States"],
    top["Adoption_percentage"]
)

plt.xticks(rotation=90)

plt.title(
    "Top States by Adoption Percentage"
)

plt.show()
```

# Governance Activity Dashboard

```python
plt.figure(figsize=(12,6))

top=df.sort_values(
    "Total_number(sum)",
    ascending=False
).head(10)

plt.bar(
    top["States"],
    top["Total_number(sum)"]
)

plt.xticks(rotation=90)

plt.title(
    "Governance Activity Dashboard"
)

plt.show()
```

# Dashboard Figure

```python
fig,ax=plt.subplots(
    1,
    2,
    figsize=(15,6)
)

top_adoption=df.sort_values(
    "Adoption_percentage",
    ascending=False
).head(10)

ax[0].bar(
    top_adoption["States"],
    top_adoption["Adoption_percentage"]
)

ax[0].set_title(
    "Top Adoption Percentage"
)

ax[0].tick_params(
    axis="x",
    rotation=90
)

top_activity=df.sort_values(
    "Total_number(sum)",
    ascending=False
).head(10)

ax[1].bar(
    top_activity["States"],
    top_activity[
        "Total_number(sum)"
    ]
)

ax[1].set_title(
    "Governance Activity"
)

ax[1].tick_params(
    axis="x",
    rotation=90
)

plt.tight_layout()

plt.show()
```

```python

# --------------------------
# Dashboard Metrics
# --------------------------

total_states=df["States"].nunique()

avg_adoption=round(
df["Adoption_percentage"].mean(),
2
)

total_activity=df[
"Total_number(sum)"
].sum()

# --------------------------
# Theme Colors
# --------------------------

BG_COLOR="#2B2B2B"

CARD_COLOR="#3A3A3A"

TEXT_COLOR="#F2F2F2"

TITLE_COLOR="white"

plt.style.use(
"dark_background"
)

fig=plt.figure(

figsize=(18,12),

facecolor=BG_COLOR

)

# --------------------------
# Dashboard Title
# --------------------------

fig.suptitle(

"E-Governance Analytics Dashboard",

fontsize=24,

fontweight="bold",

color="white"

)

fig.text(

0.5,

0.94,

"Panchayat Digital Governance Indicators Analysis",

ha="center",

fontsize=14,

color="lightgray"

)

# --------------------------
# KPI CARDS
# --------------------------

fig.text(

0.15,

0.88,

f"Total States\n{total_states}",

fontsize=16,

bbox=dict(

facecolor=CARD_COLOR,

pad=15

)

)

fig.text(

0.42,

0.88,

f"Avg Adoption %\n{avg_adoption}",

fontsize=16,

bbox=dict(

facecolor=CARD_COLOR,

pad=15

)

)

fig.text(

0.70,

0.88,

f"Governance Activity\n{total_activity}",

fontsize=16,

bbox=dict(

facecolor=CARD_COLOR,

pad=15

)

)

# --------------------------
# Adoption Chart
# --------------------------

ax1=plt.subplot(
2,
2,
1
)

top_adoption=df.sort_values(

"Adoption_percentage",

ascending=False

).head(10)

bars=ax1.bar(

top_adoption["States"],

top_adoption["Adoption_percentage"]

)

ax1.set_title(

"State-wise Adoption Analysis",

fontsize=15,

pad=15

)

ax1.set_facecolor(
CARD_COLOR
)

ax1.tick_params(

axis="x",

rotation=90,

colors="white"

)

for i,v in enumerate(

top_adoption[
"Adoption_percentage"
]

):

    ax1.text(

        i,

        v,

        round(v,1),

        ha="center",

        fontsize=8

    )

# --------------------------
# Governance Chart
# --------------------------

ax2=plt.subplot(
2,
2,
2
)

top_activity=df.sort_values(

"Total_number(sum)",

ascending=False

).head(10)

bars2=ax2.bar(

top_activity[
"States"
],

top_activity[
"Total_number(sum)"
]

)

ax2.set_title(

"Governance Activity Comparison",

fontsize=15,

pad=15

)

ax2.set_facecolor(
CARD_COLOR
)

ax2.tick_params(

axis="x",

rotation=90,

colors="white"

)

for i,v in enumerate(

top_activity[
"Total_number(sum)"
]

):

    ax2.text(

        i,

        v,

        round(v),

        ha="center",

        fontsize=8

    )

# --------------------------
# Modern Donut
# --------------------------

ax3=plt.subplot(
2,
2,
3
)

donut=top_adoption.head(
5
)

wedges,texts,autotexts=ax3.pie(

donut[
"Adoption_percentage"
],

labels=donut[
"States"
],

autopct="%1.1f%%",

pctdistance=.80,

textprops=dict(

color="white",

fontsize=10

),

wedgeprops=dict(

width=.35,

edgecolor=BG_COLOR

)

)

centre=plt.Circle(

(0,0),

0.55,

fc=BG_COLOR

)

ax3.add_artist(
centre
)

ax3.set_title(

"Top Adoption Distribution",

fontsize=15

)

# --------------------------
# Table
# --------------------------

ax4=plt.subplot(
2,
2,
4
)

ax4.axis(
"off"
)

table_data=df[
[
"States",
"Adoption_percentage",
"Total_number(sum)"
]
].head(
10
)

table=ax4.table(

cellText=table_data.values,

colLabels=table_data.columns,

loc="center"

)

table.scale(
1,
2
)

table.auto_set_font_size(
False
)

table.set_fontsize(
10
)

for key,cell in table.get_celld().items():

    cell.set_facecolor(
        CARD_COLOR
    )

    cell.set_text_props(
        color="white"
    )

ax4.set_title(

"State Comparison Table",

fontsize=15

)

# --------------------------

plt.tight_layout(
pad=3
)

plt.savefig(

"dashboard_output.png",

dpi=300,

bbox_inches="tight"

)

plt.show()
```
