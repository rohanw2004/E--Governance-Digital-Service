
### 7.DATA TYPE CONVERSION
as I have heard that Government stores the data not just in numbers but alsp in alphabets, digits, symbols. To check we will perform:

```python
numeric_columns = [
    "Gram Panchayat Development Plans Uploaded",
    "Block Panchayats Plan uploaded",
    "ZillaPanchayats Plan uploaded",
    "Total Number of Village Panchayats & Equivalent",
    "Total Number of Block Panchayats & Equivalent",
    "Total Number of ZillaPanchayats& Equivalent"
]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

### 9. FEATURE ENGINEERING
Here, No we will sum up all the required information from the data which is been used by government.

First, plans uploaded

```python
df["Plans Uploaded"] = (
    df["Gram Panchayat Development Plans Uploaded"] +
    df["Block Panchayats Plan uploaded"] +
    df["ZillaPanchayats Plan uploaded"]
)
```

Then, Total number of blocks

```python
df["Total number(sum)"] = (
    df["Total Number of Village Panchayats & Equivalent"] +
    df["Total Number of Block Panchayats & Equivalent"] +
    df["Total Number of ZillaPanchayats& Equivalent"]
)
```

Now, We are Going to rationalize the data divisional information based on project.

```python
df["Adoption percentage = (plans uploaded/ Total Panchayats)*100"] = (
    df["Plans Uploaded"] /
    df["Total number(sum)"]
) * 100
```
